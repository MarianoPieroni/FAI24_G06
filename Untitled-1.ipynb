Desenvolver um sistema preditivo inteligente capaz de prever o desempenho académico de estudantes com base nos seus hábitos e características pessoais,

- Prever desempenho académico baseado em hábitos dos alunos(0-25, 26-50, 51-70, 71-100 pontos)
- Identificar fatores críticos para sucesso académico

## tratamento de valores nulos e duplicados


Valores nulos são dados não registrados no dataset,para tratar este valores pode-se fazer de 3 maneiras:
- Remoção
- Imputação Numérica
- Imputação Categórica

- Nao se usa remoçao porque cada linha representa um aluno e como teria-se de remover uma grande parte,poderia-se tar a remover um grupo especifico perdendo padroes importantes para a amostra final.
- Para variaveis numericas usou-se a Mediana em vez da media por que se ouver outliers a media pode ser afetada enquanto mediana nao, representa melhor o valor"tipico".
- para variaveis categoricas usou-se a moda porque mantem a distribuiçao natural e causa um baixo impacto na analise.


```python
# tratamento de valores nulos
if df.isnull().sum().sum() > 0:
     # Preencher numéricos com mediana
 numeric_cols = df.select_dtypes(include=[np.number]).columns
 df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
    # Preencher categóricos com moda
 categorical_cols = df.select_dtypes(include=['object']).columns
 for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0]) 

print(f"\n-valores nulos apos tratamento:{df.isnull().sum().sum()}")
```

### 1.2-Preparação dos Dados 


```python
print("PREPARACAO DOS DADOS")
print("="*50)

# Criar variável target 'Desempenho'
df['Desempenho'] = pd.cut(df['exam_score'],
                         bins=[0, 25, 50, 70, 100],
                         labels=['0-25', '26-50', '51-70', '71-100'],
                         include_lowest=True,
                         right=True)
# Distribuição da variável target
distribuicao = df['Desempenho'].value_counts().sort_index()
print("Distribuiçao do Desempenho:")
print(distribuicao)

# Visualização(novo para o notebook)
plt.figure(figsize=(6, 6))
sns.countplot(data=df, x='Desempenho', palette='viridis')
plt.title('Distribuiçao do Desempenho dos Estudantes')
plt.xlabel('Faixa de Desempenho')
plt.ylabel('Numero de Estudantes')
plt.show()
```

### 1.3-Encoding de Variáveis Categóricas


- Selecionou-se todas variaveis categoricas, atraves do labelEncoder atribui valores interios as opçoes de cada variavel para depois serem aplicadas nos algoritmos:

- gender_encoded-[0 1 2]
- diet_quality_encoded-[0 1 2]
- parental_education_level_encoded-[2 1 0 3]
- internet_quality_encoded-[0 2 1]
- extracurricular_participation_encoded-[1 0]
- part_time_job_encoded-[0 1]


```python
print("ENCODING DE VARIAVEIS CATEGORICAS")
print("="*50)

categorical_columns = ['gender', 'diet_quality', 'parental_education_level',
                      'internet_quality', 'extracurricular_participation', 'part_time_job']

for col in categorical_columns:
    if col in df.columns:
        encoded_col = col + '_encoded'
        le = LabelEncoder()
        df[encoded_col] = le.fit_transform(df[col].astype(str))
        print(f"{col} -> {encoded_col}")

encoded_cols = [col for col in df.columns 
                if col.endswith('_encoded')and col in df.columns]
print(f"\nColunas codificadas criadas: {len(encoded_cols)}")
print(encoded_cols)

#VERIFICAÇAO DOS VALORES UNICOS DAS COLUNAS ENCODED(novo para o notebook)
print(f"\ngender_encoded: {df['gender'].unique()} -> {df['gender_encoded'].unique()}")
print(f"diet_quality_encoded: {df['diet_quality'].unique()} -> {df['diet_quality_encoded'].unique()}")
print(f"parental_education_level_encoded: {df['parental_education_level'].unique()} -> {df['parental_education_level_encoded'].unique()}")
print(f"internet_quality_encoded: {df['internet_quality'].unique()} -> {df['internet_quality_encoded'].unique()}")
print(f"extracurricular_participation_encoded: {df['extracurricular_participation'].unique()} -> {df['extracurricular_participation_encoded'].unique()}")
print(f"part_time_job_encoded: {df['part_time_job'].unique()} -> {df['part_time_job_encoded'].unique()}")
```

### 1.4-Seleção de Features e Split dos Dados

- Seleciona-se os features importantes para o modelo, por exemplo nao se usa o id de estudante por nao ser uma variavel preditivel, e nao se usa o exam_score porque ja ta a ser usado para criar target

- Divide-se os dados em conjuntos de treino e teste para:
   - Treino: Ensinar o modelo a reconhecer padrões
   - Teste: Avaliar como o modelo performa com dados nunca vistos

- Usamos o stratify porque:
   -  Garante que todos os níveis de desempenho estão representados
   -  Evita que uma classe fique só no treino ou só no teste
   -  Modelo aprende e é testado em todas as categorias



```python
print("SELEÇAO DE FEATURES E DIVISAO DOS DADOS")
print("="*50)

final_features = [
    'age', 'study_hours_per_day', 'social_media_hours', 'netflix_hours',
    'attendance_percentage', 'sleep_hours', 'exercise_frequency', 'mental_health_rating',
] + encoded_cols

print(f"Total de features selecionadas: {len(final_features)}")
print(f"- Numericas: {len([f for f in final_features if 'encoded' not in f])}")
print(f"- Categoricas codificadas: {len([f for f in final_features if 'encoded' in f])}")
print("\nFeatures selecionadas:")
print(final_features)

# Preparar dados para modelaçao
X = df[final_features]
y = df['Desempenho']

# Split dos dados
x_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, #20% para teste, 80% para treino
      random_state=42,   #Reprodutibilidade (mesma divisão sempre)
        stratify=y       #Mantém proporções iguais em treino/teste
)

print(f"\nDivisão dos dados:")
print(f"- Treino: {x_train.shape[0]} amostras")
print(f"- Teste: {X_test.shape[0]} amostras")
print(f"- Features: {x_train.shape[1]}")
```

### 1.5-Normalização dos Dados e Mapeamento para Relatório



- Normaliza-se os dados porque os K-NN são sensíveis à escala dos dados e normalização garante que todas as features contribuam igualmente.

- K-NN usa dados normalizados, Random Forest usa dados originais

```python
print("NORMALIZAÇÃO DOS DADOS")
print("="*50)

scaler = StandardScaler()
x_trainscaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(X_test)

print("Dados normalizados para K-NN")
print(f"- X_train_scaled shape: {x_trainscaled.shape}")
print(f"- X_test_scaled shape: {X_test_scaled.shape}")

faixa_para_intervalo = {
    '0-25': '0-25 pontos',
    '26-50': '26-50 pontos',
    '51-70': '51-70 pontos', 
    '71-100': '71-100 pontos'
}

print("\nMapeamento faixa -> intervalo:")
for faixa, intervalo in faixa_para_intervalo.items():
    print(f"  {faixa} -> {intervalo}")
```

## 2.1-Random Forest com Hyperparameter Tuning

```python
print("RANDOM FOREST COM HYPERPARAMETER TUNING")
print("="*50)

# Definir grid de parâmetros
rf_param_grid = {
        'n_estimators': range(50, 301),
        'max_depth': range(5, 21),
        'min_samples_split': range(2, 16),
        'min_samples_leaf': range(1, 6),
        'bootstrap': [True, False]
}


rf_search = RandomizedSearchCV(
        RandomForestClassifier(random_state=42),
        rf_param_grid,
        n_iter=100,
        cv=5,
        scoring='accuracy',
        n_jobs=-1,
        random_state=42
    )
rf_search.fit(x_train, y_train)
print(f"Melhores parametros: {rf_search.best_params_}")
print(f"Melhor accuracy (CV): {rf_search.best_score_:.3f}")

# Modelo final com melhores parâmetros(novo para o notebook)
rf_model = rf_search.best_estimator_
y_pred_rf = rf_model.predict(X_test)
acc_rf = accuracy_score(y_test, y_pred_rf)

print(f"Accuracy no teste: {acc_rf:.3f}")
```

## 2.2-K-NN com Hyperparameter Tuning

```python
print("K-NN COM HYPERPARAMETER TUNING")
print("="*50)

# Definir grid de parâmetros
knn_param_grid = {
    'n_neighbors':  range(3, 31),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

knn_search = RandomizedSearchCV(
        KNeighborsClassifier(),
        knn_param_grid,
        n_iter=50,
        cv=5,
        scoring='accuracy',
        n_jobs=-1,
        random_state=42
    )
knn_search.fit(x_trainscaled, y_train)
print(f"Melhores parametros: {knn_search.best_params_}")
print(f"Melhor accuracy (CV): {knn_search.best_score_:.3f}")

# Modelo final com melhores parâmetros(novo para o notebook)
knn_model = knn_search.best_estimator_
y_pred_knn = knn_model.predict(X_test_scaled)
acc_knn = accuracy_score(y_test, y_pred_knn)

print(f"Accuracy no teste: {acc_knn:.3f}")
```

## 3.1-Sistema de Previsão

```python
def load_trained_models():
    try:
        rf_model = joblib.load('../models/random_forest_model.pkl')
        knn_model = joblib.load('../models/knn_model.pkl')
        scaler = joblib.load('../models/scaler.pkl')
        return rf_model, knn_model, scaler
    except:
        return None, None, None
    
def predict_student_performance(student_data, rf_model, knn_model, scaler, faixa_para_intervalo):
    # Converter para DataFrame se for dict
    if isinstance(student_data, dict):
        student_df = pd.DataFrame([student_data])
    else:
        student_df = student_data
    
    # Previsão Random Forest
    faixa_rf = rf_model.predict(student_df)[0]
    
    # Previsão K-NN (precisa de scaling)
    student_scaled = scaler.transform(student_df)
    faixa_knn = knn_model.predict(student_scaled)[0]
    
    # Verificar concordância
    concordancia = faixa_rf == faixa_knn
    confianca = "Alta" if concordancia else "Media"
    
    # Resultado final (usar RF se concordam, ou ambos se discordam)
    if concordancia:
        faixa_final = faixa_rf
    else:
        faixa_final = f"{faixa_rf}/{faixa_knn}"
    
    return {
        'faixa_prevista': faixa_final,
        'intervalo': faixa_para_intervalo.get(faixa_rf, 'N/A'),
        'confianca': confianca,
        'random_forest': faixa_rf,
        'knn': faixa_knn,
        'concordancia': concordancia
    }

def show_individual_predictions(prediction, student_name="Aluno"):

    print("="*50)
    print(f"PREVISOES INDIVIDUAIS DOS MODELOS")
    print("="*50)
    print(f"Estudante: {student_name}")
    print(f"Random Forest: {prediction['random_forest']}")
    print(f"K-NN: {prediction['knn']}")
    print(f"Concordancia: {' SIM' if prediction['concordancia'] else ' NAO'}")
    print("="*50)

def generate_report(prediction, student_name="Aluno"):

    print("\n" + "="*50)
    print(f"RELATORIO FINAL - PREVISAO COMBINADA")
    print("="*50)
    print(f"Estudante: {student_name}")
    print(f"Desempenho previsto: {prediction['faixa_prevista']}")
    print(f"Intervalo de nota: {prediction['intervalo']}")
    print(f"Nivel de confianca: {prediction['confianca']}")
    print("="*50)
    
    # Recomendações
    recomendacoes = {
        '0-25': "Apoio URGENTE",
        '26-50': "Acompanhamento proximo", 
        '51-70': "Orientaçao de estudo",
        '71-100': "Manter excelencia"
    }
    
    faixa_principal = prediction['random_forest']
    if faixa_principal in recomendacoes:
        print(f"RECOMENDACAO: {recomendacoes[faixa_principal]}")

def generate_complete_prediction_report(student_data, rf_model, knn_model, scaler, faixa_para_intervalo, student_name="Aluno"):

    # Fazer previsão
    prediction = predict_student_performance(
        student_data, rf_model, knn_model, scaler, faixa_para_intervalo
    )
    
    # Mostrar previsões individuais
    show_individual_predictions(prediction, student_name)
    
    # Mostrar resultado combinado
    generate_report(prediction, student_name)
    
    return prediction
```

## Análise de Clustering - K-Means

### Importações


```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score  #reponsavel pela avaliaçao da distancia do kmeans
```

### Função de Otimização do Número de Clusters

```python
def encontrar_melhor_k(dados_normalizados, max_k=8):

    print("\nHYPERPARAMETER OPTIMIZATION")
    print("=" * 30)
    
    resultados = []
    
    print("Testando valores de K:")
    for k in range(2, max_k + 1):
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        clusters = kmeans.fit_predict(dados_normalizados)
        
        # Calcular métricas
        wcss = kmeans.inertia_
        silhouette = silhouette_score(dados_normalizados, clusters)
        
        resultados.append({
            'k': k,
            'wcss': wcss,
            'silhouette': silhouette
        })
        
        print(f"K={k}: WCSS = {wcss:8.2f} | Silhouette = {silhouette:.3f}")
    
    # Encontrar melhor K baseado no silhouette
    melhor_resultado = max(resultados, key=lambda x: x['silhouette'])
    melhor_k = melhor_resultado['k']
    
    print(f"\nMELHOR K: {melhor_k}")
    print(f"Silhouette: {melhor_resultado['silhouette']:.3f}")
    
    return melhor_k

```


### Análise e Caracterização dos Clusters
- Carregamento e Preparação dos Dados
- Aplicação do Algoritmo K-Means


```python
def analisar_clusters_kmeans():
    # Carregar dados
    try:
        df = pd.read_csv('student_habits_performance.csv')
        #print(f"Dataset carregado {df.shape[0]}")
    except:
        print("Arquivo não encontrado")
        return None, None
    
    # Preparando dados para clustering
    features = ['study_hours_per_day', 'social_media_hours', 'sleep_hours', 
                'exercise_frequency', 'attendance_percentage']
    
    dados_clustering = df[features].copy()
    scaler = StandardScaler()
    dados_normalizados = scaler.fit_transform(dados_clustering) # Normalização

    k_otimo = encontrar_melhor_k(dados_normalizados)

    #mudar aqui para ter mais clusters
 #   kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
    #otimizando
    kmeans = KMeans(n_clusters=k_otimo, init='k-means++', random_state=42, n_init=10)
    #fit_perdict = funçao principal do kmeans onde faz os 3 passos 
    clusters = kmeans.fit_predict(dados_normalizados)
    
    df_com_clusters = df.copy()
    df_com_clusters['Cluster'] = clusters

        # Estatísticas por cluster
    for cluster in sorted(df_com_clusters['Cluster'].unique()):
        cluster_data = df_com_clusters[df_com_clusters['Cluster'] == cluster]
        
        print(f"\nCLUSTER {cluster + 1} ({len(cluster_data)} estudantes):")
        print(f"Notas: {cluster_data['exam_score'].mean():.1f}")
        print(f"Horas estudo/dia: {cluster_data['study_hours_per_day'].mean():.1f}h")
        print(f"Horas no insta: {cluster_data['social_media_hours'].mean():.1f}h")
        print(f"Horas sono: {cluster_data['sleep_hours'].mean():.1f}h")
        print(f"Presença: {cluster_data['attendance_percentage'].mean():.1f}%")
    
    return kmeans, df_com_clusters
```

### Função Principal com Tratamento de Erros

```python
def unsupervised_():
    try:
        modelo_kmeans, dados_clusters = analisar_clusters_kmeans()
        return modelo_kmeans, dados_clusters
    except Exception as e:
        print(f"Erro na análise: {e}")
        return None, None
```

# Association Rules Analysis - Apriori Algorithm

## Objetivo principal
Descobrir padrões de associação entre os hábitos dos estudantes e seus clusters, identificando combinações de comportamentos que caracterizam cada grupo através do algoritmo Apriori.

### Importações

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
from unsupervised import analisar_clusters_kmeans
```

## Função Principal

### 1. **Carregamento e Inicialização**
- Executa o algoritmo de clustering para obter dados segmentados
- Verifica disponibilidade dos dados para análise
- Prepara o ambiente para o Association Rules

### 2. **Preparação e Discretização de Dados**  
- Cria cópia segura dos dados originais
- Transforma variáveis numéricas contínuas em categorias
- Converte horas de estudo e redes sociais em níveis (baixo/médio/alto)

### 3. **Criação de Transações**
- Converte cada estudante em uma "transação" de mercado
- Formata todos os hábitos como itens categóricos
- Inclui o cluster como item alvo para análise

### 4. **Conversão para Formato Binário**
- Transforma transações em matriz binária (one-hot encoding)
- Prepara dados no formato exigido pelo algoritmo Apriori
- Identifica todos os itens únicos do dataset

### 5. **Otimização de Parâmetros**
- Testa sistematicamente combinações de min_support e min_confidence
- Avalia a quantidade de regras relevantes geradas
- Seleciona automaticamente os melhores parâmetros

### 6. **Aplicação do Algoritmo Apriori**
- Executa o Apriori com os parâmetros otimizados
- Gera itemsets frequentes e regras de associação
- Filtra regras com lift significativo (> 1.6)

### 7. **Análise por Cluster**
- Agrupa regras por cluster de destino
- Remove redundâncias (cluster nos antecedentes)
- Apresenta regras no formato "SE [hábitos] ENTÃO [cluster]"

### 8. **Métricas de Performance**
- Calcula estatísticas de qualidade das regras
- Analisa suporte, confiança e lift médios
- Identifica regras com alta confiança (>70%)

```python
def association_rules_():
    
    # Carregar dados(o (modelo_kmeans) nao esta sendo usado pois so precisamos do df, e a funçao retorna 2 valores por isso a colocamos)
    modelo_kmeans, df_com_clusters = analisar_clusters_kmeans()

    print("\nAlgoritmo Apriori")
    if df_com_clusters is None:
        print("não encontrados")
        return None
    
    print(f"dados: {df_com_clusters.shape[0]} estudantes")
    
    #preparar os dados
    # copia para nao modificar os dados originais
    df_preparado = df_com_clusters.copy()
    df_preparado['estudo_cat'] = pd.cut(df_preparado['study_hours_per_day'], 
                                      bins=[0, 2, 4, 24], 
                                      labels=['baixo', 'medio', 'alto'])
    
    df_preparado['redes_cat'] = pd.cut(df_preparado['social_media_hours'], 
                                     bins=[0, 2, 4, 24], 
                                     labels=['baixo', 'medio', 'alto'])
    
    # prepara os dados de cada aluno para o apriori analizar
    transacoes = []
    for _, estudante in df_preparado.iterrows():
        transacao = [
            f"Estudo_{estudante['estudo_cat']}",
            f"RedesSociais_{estudante['redes_cat']}",
            f"Sono_{estudante['sleep_hours']:.0f}h",
            f"Exercicio_{estudante['exercise_frequency']}",
            f"Trabalho_{estudante['part_time_job']}",
            f"Cluster_{estudante['Cluster'] + 1}"
        ]
        transacoes.append(transacao)
    
    # Converter para binário
    te = TransactionEncoder()
    te_array = te.fit(transacoes).transform(transacoes)
    df_binario = pd.DataFrame(te_array, columns=te.columns_)
    
    print(f"Transações criadas: {len(transacoes)}")
    
    #variáveis que armazena os melhores resultados
    melhor_rules = None
    melhor_combinacao = None
    
    #testa 2 valores de suporte mínimo
    for min_support in [0.05, 0.08]:
        #para cada suporte, testa 2 valores de confiança mínima
        for min_confidence in [0.6, 0.7]:

        # Aplica algoritmo Apriori para encontrar itemsets frequentes
            frequent_itemsets = apriori(df_binario, 
                                      min_support=min_support, #a frequencia minima que um itemset deve ter
                                      use_colnames=True) #usa os nomes das colunas em vez de índices
        # Verifica se encontrou algum itemset frequente    
            if len(frequent_itemsets) > 0:
                #Gera regras de associação a partir dos itemsets frequentes
                rules = association_rules(frequent_itemsets, 
                                        metric="confidence", #usa confiança como métrica para filtrar
                                        min_threshold=min_confidence) #confiança mínima que uma regra deve ter
                
                if len(rules) > 0:
                    regras_boas = rules[rules['lift'] > 1.2]
                    # Atualiza as melhores regras se é a primeira combinação testada ou se encontrou regras melhores
                    if melhor_rules is None or len(regras_boas) > len(melhor_rules):
                        melhor_rules = regras_boas
                        melhor_combinacao = (min_support, min_confidence)
    
    print(f"melhores parametros: support={melhor_combinacao[0]}, confidence={melhor_combinacao[1]}")
    
    # refaz o algoritmo Apriori com melhores parâmetros
    frequent_itemsets = apriori(df_binario, 
                              min_support=melhor_combinacao[0], 
                              use_colnames=True)
    #gera as regras definitivas
    rules = association_rules(frequent_itemsets, 
                            metric="confidence", 
                            min_threshold=melhor_combinacao[1])
    
    #INTERMEDIATE AND FINAL RESULTS
    print(f"\nINTERMEDIATE AND FINAL RESULTS")
    print("-" * 40)
    print(f"Itemsets frequentes: {len(frequent_itemsets)}")
    print(f"Regras geradas: {len(rules)}")
    
    # Filtrar regras relevantes
    regras_relevantes = rules[rules['lift'] > 1.6]
    print(f"Regras relevantes (lift > 1.2): {len(regras_relevantes)}")
    
    # Análise por cluster
    print(f"\nANÁLISE POR CLUSTER:")
    print("=" * 50)
    
    #percorre todos os clusters
    for cluster_num in sorted(df_preparado['Cluster'].unique()):
        print(f"\nCLUSTER {cluster_num + 1}") 
        
        #filtra as regras relevantes XX
        regras_cluster = regras_relevantes[regras_relevantes['consequents'].apply(
                lambda x: f"Cluster_{cluster_num + 1}" in str(x)
            )
        ]
        #verifica se encontrou regras
        if len(regras_cluster) > 0:
            # Percorre cada regra do cluster
            for i, (_, regra) in enumerate(regras_cluster.iterrows()):
                #Para cada item nos antecedentes da regra, mantem apenas os q n começam com celula
                antecedentes = [str(x) for x in regra['antecedents'] if not x.startswith('Cluster_')]
                #printa se for valido
                if antecedentes:
                    print(f"  {i+1}. SE {antecedentes}")
                    print(f"     Conf: {regra['confidence']:.1%} | Lift: {regra['lift']:.2f}")
        else:
            print("Nenhuma regra boa")
    
    #PERFORMANCE METRICS ANALYSIS
    print(f"\nanalise de performace")
    print("-" * 40)
    
    if len(regras_relevantes) > 0:
        print(f"metricas das {len(regras_relevantes)} regras relevantes:")
        print(f"Suporte médio: {regras_relevantes['support'].mean():.3f}")
        print(f"Confiança média: {regras_relevantes['confidence'].mean():.3f}")
        print(f"Lift médio: {regras_relevantes['lift'].mean():.3f}")
        
        # Uma análise extra essencial
        alta_confianca = len(regras_relevantes[regras_relevantes['confidence'] > 0.7])
        print(f"Regras com alta confiança (>70%): {alta_confianca}")
    
    return regras_relevantes, df_preparado
```
