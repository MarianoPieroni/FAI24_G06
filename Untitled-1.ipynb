```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV,RandomizedSearchCV
import joblib
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
print("Bibliotecas importadas com sucesso!")

```

### 1.-CARREGAR E EXPLORAR DADOS


```python
df = pd.read_csv('student_habits_performance.csv')

print("Dataset")
print(f"\n-Dimensoes:{df.shape}")
print(f"\n-Colunas:{df.columns}")
print(f"\n-valores nulos:{df.isnull().sum().sum()}")
print(f"\n-duplicados:{df.duplicated().sum()}")

#novo para o notebook
print("\nPrimeiras 5 linhas:")
display(df.head())
```

### 1.1-Análise Exploratória - Estatísticas Descritivas



```python
#novo para o notebook
print("ESTATISTICAS DESCRITIVAS")
print("="*50)

print("\nEstatisticas das variaveis numericas:")
display(df.describe())

print("\nTipos de dados:")
print(df.dtypes)
```

## tratamento de valores nulos e duplicados


Valores nulos são dados não registrados no dataset,para tratar este valores pode-se fazer de 3 maneiras:
- Remoção
- Imputação Numérica
- Imputação Categórica

- Nao se usa remoçao porque cada linha representa um aluno e como teria-se de remover uma grande parte,poderia-se tar a remover um grupo especifico perdendo padroes importantes para a amostra final.
- Para variaveis numericas usou-se a Mediana em vez da media por que se ouver outliers a media pode ser afetada enquanto mediana nao, representa melhor o valor"tipico".
- para variaveis categoricas usou-se a moda porque mantem a distribuiçao natural e causa um baixo impacto na analise.


```python
# tratamento de valores nulos
if df.isnull().sum().sum() > 0:
     # Preencher numéricos com mediana
 numeric_cols = df.select_dtypes(include=[np.number]).columns
 df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
    # Preencher categóricos com moda
 categorical_cols = df.select_dtypes(include=['object']).columns
 for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0]) 

print(f"\n-valores nulos apos tratamento:{df.isnull().sum().sum()}")
```

### 1.2-Preparação dos Dados 


```python
print("PREPARACAO DOS DADOS")
print("="*50)

# Criar variável target 'Desempenho'
df['Desempenho'] = pd.cut(df['exam_score'],
                         bins=[0, 25, 50, 70, 100],
                         labels=['0-25', '26-50', '51-70', '71-100'],
                         include_lowest=True,
                         right=True)
# Distribuição da variável target
distribuicao = df['Desempenho'].value_counts().sort_index()
print("Distribuiçao do Desempenho:")
print(distribuicao)

# Visualização(novo para o notebook)
plt.figure(figsize=(6, 6))
sns.countplot(data=df, x='Desempenho', palette='viridis')
plt.title('Distribuiçao do Desempenho dos Estudantes')
plt.xlabel('Faixa de Desempenho')
plt.ylabel('Numero de Estudantes')
plt.show()
```

### 1.3-Encoding de Variáveis Categóricas


- Selecionou-se todas variaveis categoricas, atraves do labelEncoder atribui valores interios as opçoes de cada variavel para depois serem aplicadas nos algoritmos:

- gender_encoded-[0 1 2]
- diet_quality_encoded-[0 1 2]
- parental_education_level_encoded-[2 1 0 3]
- internet_quality_encoded-[0 2 1]
- extracurricular_participation_encoded-[1 0]
- part_time_job_encoded-[0 1]


```python
print("ENCODING DE VARIAVEIS CATEGORICAS")
print("="*50)

categorical_columns = ['gender', 'diet_quality', 'parental_education_level',
                      'internet_quality', 'extracurricular_participation', 'part_time_job']

for col in categorical_columns:
    if col in df.columns:
        encoded_col = col + '_encoded'
        le = LabelEncoder()
        df[encoded_col] = le.fit_transform(df[col].astype(str))
        print(f"{col} -> {encoded_col}")

encoded_cols = [col for col in df.columns 
                if col.endswith('_encoded')and col in df.columns]
print(f"\nColunas codificadas criadas: {len(encoded_cols)}")
print(encoded_cols)

#VERIFICAÇAO DOS VALORES UNICOS DAS COLUNAS ENCODED(novo para o notebook)
print(f"\ngender_encoded: {df['gender'].unique()} -> {df['gender_encoded'].unique()}")
print(f"diet_quality_encoded: {df['diet_quality'].unique()} -> {df['diet_quality_encoded'].unique()}")
print(f"parental_education_level_encoded: {df['parental_education_level'].unique()} -> {df['parental_education_level_encoded'].unique()}")
print(f"internet_quality_encoded: {df['internet_quality'].unique()} -> {df['internet_quality_encoded'].unique()}")
print(f"extracurricular_participation_encoded: {df['extracurricular_participation'].unique()} -> {df['extracurricular_participation_encoded'].unique()}")
print(f"part_time_job_encoded: {df['part_time_job'].unique()} -> {df['part_time_job_encoded'].unique()}")
```

### 1.4-Seleção de Features e Split dos Dados

- Seleciona-se os features importantes para o modelo, por exemplo nao se usa o id de estudante por nao ser uma variavel preditivel, e nao se usa o exam_score porque ja ta a ser usado para criar target

- Divide-se os dados em conjuntos de treino e teste para:
   - Treino: Ensinar o modelo a reconhecer padrões
   - Teste: Avaliar como o modelo performa com dados nunca vistos

- Usamos o stratify porque:
   -  Garante que todos os níveis de desempenho estão representados
   -  Evita que uma classe fique só no treino ou só no teste
   -  Modelo aprende e é testado em todas as categorias



```python
print("SELEÇAO DE FEATURES E DIVISAO DOS DADOS")
print("="*50)

final_features = [
    'age', 'study_hours_per_day', 'social_media_hours', 'netflix_hours',
    'attendance_percentage', 'sleep_hours', 'exercise_frequency', 'mental_health_rating',
] + encoded_cols

print(f"Total de features selecionadas: {len(final_features)}")
print(f"- Numericas: {len([f for f in final_features if 'encoded' not in f])}")
print(f"- Categoricas codificadas: {len([f for f in final_features if 'encoded' in f])}")
print("\nFeatures selecionadas:")
print(final_features)

# Preparar dados para modelaçao
X = df[final_features]
y = df['Desempenho']

# Split dos dados
x_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, #20% para teste, 80% para treino
      random_state=42,   #Reprodutibilidade (mesma divisão sempre)
        stratify=y       #Mantém proporções iguais em treino/teste
)

print(f"\nDivisão dos dados:")
print(f"- Treino: {x_train.shape[0]} amostras")
print(f"- Teste: {X_test.shape[0]} amostras")
print(f"- Features: {x_train.shape[1]}")
```

### 1.5-Normalização dos Dados e Mapeamento para Relatório



- Normaliza-se os dados porque os K-NN são sensíveis à escala dos dados e normalização garante que todas as features contribuam igualmente.

- K-NN usa dados normalizados, Random Forest usa dados originais

```python
print("NORMALIZAÇÃO DOS DADOS")
print("="*50)

scaler = StandardScaler()
x_trainscaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(X_test)

print("Dados normalizados para K-NN")
print(f"- X_train_scaled shape: {x_trainscaled.shape}")
print(f"- X_test_scaled shape: {X_test_scaled.shape}")

faixa_para_intervalo = {
    '0-25': '0-25 pontos',
    '26-50': '26-50 pontos',
    '51-70': '51-70 pontos', 
    '71-100': '71-100 pontos'
}

print("\nMapeamento faixa -> intervalo:")
for faixa, intervalo in faixa_para_intervalo.items():
    print(f"  {faixa} -> {intervalo}")
```

## 2.1-Random Forest com Hyperparameter Tuning

```python
print("RANDOM FOREST COM HYPERPARAMETER TUNING")
print("="*50)

# Definir grid de parâmetros
rf_param_grid = {
        'n_estimators': range(50, 301),
        'max_depth': range(5, 21),
        'min_samples_split': range(2, 16),
        'min_samples_leaf': range(1, 6),
        'bootstrap': [True, False]
}


rf_search = RandomizedSearchCV(
        RandomForestClassifier(random_state=42),
        rf_param_grid,
        n_iter=100,
        cv=5,
        scoring='accuracy',
        n_jobs=-1,
        random_state=42
    )
rf_search.fit(x_train, y_train)
print(f"Melhores parametros: {rf_search.best_params_}")
print(f"Melhor accuracy (CV): {rf_search.best_score_:.3f}")

# Modelo final com melhores parâmetros(novo para o notebook)
rf_model = rf_search.best_estimator_
y_pred_rf = rf_model.predict(X_test)
acc_rf = accuracy_score(y_test, y_pred_rf)

print(f"Accuracy no teste: {acc_rf:.3f}")
```

## 2.2-K-NN com Hyperparameter Tuning

```python
print("K-NN COM HYPERPARAMETER TUNING")
print("="*50)

# Definir grid de parâmetros
knn_param_grid = {
    'n_neighbors':  range(3, 31),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

knn_search = RandomizedSearchCV(
        KNeighborsClassifier(),
        knn_param_grid,
        n_iter=50,
        cv=5,
        scoring='accuracy',
        n_jobs=-1,
        random_state=42
    )
knn_search.fit(x_trainscaled, y_train)
print(f"Melhores parametros: {knn_search.best_params_}")
print(f"Melhor accuracy (CV): {knn_search.best_score_:.3f}")

# Modelo final com melhores parâmetros(novo para o notebook)
knn_model = knn_search.best_estimator_
y_pred_knn = knn_model.predict(X_test_scaled)
acc_knn = accuracy_score(y_test, y_pred_knn)

print(f"Accuracy no teste: {acc_knn:.3f}")
```

## 3.1-Sistema de Previsão

```python
def load_trained_models():
    try:
        rf_model = joblib.load('../models/random_forest_model.pkl')
        knn_model = joblib.load('../models/knn_model.pkl')
        scaler = joblib.load('../models/scaler.pkl')
        return rf_model, knn_model, scaler
    except:
        return None, None, None
    
def predict_student_performance(student_data, rf_model, knn_model, scaler, faixa_para_intervalo):
    # Converter para DataFrame se for dict
    if isinstance(student_data, dict):
        student_df = pd.DataFrame([student_data])
    else:
        student_df = student_data
    
    # Previsão Random Forest
    faixa_rf = rf_model.predict(student_df)[0]
    
    # Previsão K-NN (precisa de scaling)
    student_scaled = scaler.transform(student_df)
    faixa_knn = knn_model.predict(student_scaled)[0]
    
    # Verificar concordância
    concordancia = faixa_rf == faixa_knn
    confianca = "Alta" if concordancia else "Media"
    
    # Resultado final (usar RF se concordam, ou ambos se discordam)
    if concordancia:
        faixa_final = faixa_rf
    else:
        faixa_final = f"{faixa_rf}/{faixa_knn}"
    
    return {
        'faixa_prevista': faixa_final,
        'intervalo': faixa_para_intervalo.get(faixa_rf, 'N/A'),
        'confianca': confianca,
        'random_forest': faixa_rf,
        'knn': faixa_knn,
        'concordancia': concordancia
    }

def show_individual_predictions(prediction, student_name="Aluno"):

    print("="*50)
    print(f"PREVISOES INDIVIDUAIS DOS MODELOS")
    print("="*50)
    print(f"Estudante: {student_name}")
    print(f"Random Forest: {prediction['random_forest']}")
    print(f"K-NN: {prediction['knn']}")
    print(f"Concordancia: {' SIM' if prediction['concordancia'] else ' NAO'}")
    print("="*50)

def generate_report(prediction, student_name="Aluno"):

    print("\n" + "="*50)
    print(f"RELATORIO FINAL - PREVISAO COMBINADA")
    print("="*50)
    print(f"Estudante: {student_name}")
    print(f"Desempenho previsto: {prediction['faixa_prevista']}")
    print(f"Intervalo de nota: {prediction['intervalo']}")
    print(f"Nivel de confianca: {prediction['confianca']}")
    print("="*50)
    
    # Recomendações
    recomendacoes = {
        '0-25': "Apoio URGENTE",
        '26-50': "Acompanhamento proximo", 
        '51-70': "Orientaçao de estudo",
        '71-100': "Manter excelencia"
    }
    
    faixa_principal = prediction['random_forest']
    if faixa_principal in recomendacoes:
        print(f"RECOMENDACAO: {recomendacoes[faixa_principal]}")

def generate_complete_prediction_report(student_data, rf_model, knn_model, scaler, faixa_para_intervalo, student_name="Aluno"):

    # Fazer previsão
    prediction = predict_student_performance(
        student_data, rf_model, knn_model, scaler, faixa_para_intervalo
    )
    
    # Mostrar previsões individuais
    show_individual_predictions(prediction, student_name)
    
    # Mostrar resultado combinado
    generate_report(prediction, student_name)
    
    return prediction
```
